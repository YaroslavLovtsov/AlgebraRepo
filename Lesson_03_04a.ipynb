__5*.__ Написать на Python функцию для перемножения двух произвольных матриц, не используя NumPy.

```python
def get_shape(matr):
  width = 0
  height = len(matr)
  
  if height > 0:
    width = len(matr[0])
    for inx in range(1,height):
      if len(matr[inx]) != width:
        print(width, len(matr[inx]))
        width = None
        break

  if width is None:
    raise "IncorrectSize"
  else:
    return height,width

def dot(matr1, matr2):
  try:
    h1,w1 = get_shape(matr1)
    h2,w2 = get_shape(matr2)

    if w1 != h2:
      print(w1, h2)
      print('Произведения матриц не существует')
      return None
    else:
      prod_matr = []
      for indh in range(h1):
        current_str = []
        for indw in range(w2):
          current_el = 0
          for ind_main in range(w1):
            current_el += matr1[indh][ind_main] * matr2[ind_main][indw]
          current_str.append(current_el)  
        prod_matr.append(current_str)
      return prod_matr
  except:
    print('одна из матриц задана некорректно')
    return None

matr1 = [[4,1],[5,-2],[2,3]]
matr2 = [[4,5,2],[1,-2,3]]

matr_prod = dot(matr1,matr2)
if matr_prod is not None:
  print(matr_prod)
```

 2 & 2\\
