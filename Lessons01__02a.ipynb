##**Урок 1.**

__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

**Решение:**

$$1 \cdot f_{1}(x)  + 1 \cdot f_{2}(x) + (-1) \cdot f_{3}(x) + 1 \cdot 
f_{4}(x) = 0 \;\;\; \forall x $$

Таким образом, функции линейно зависимы

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

**Решение**

$$ (-1) \cdot f_{1}(x) + (-4) \cdot f_{2}(x) + (-2) \cdot f_{3}(x) + 2 \cdot f_{4}(x) = 0\;\;\; \forall x  $$

Таким образом, функции линейно зависимы

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

**Решение**

$$ x = \frac{1}{2}\cdot b_1 + b_2 + b_3, ~ x = \Biggl(\frac{1}{2},0,0\biggr) \in L_{(b_1,b_2,b_3)}$$

__4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

а) в базисе $1$, $x$, $x^{2}$;

б) в базисе $x^{2}$, $x-1$, $1$.

**Решение**

Пусть $f(x) = 3x^{2}-2x+2 $

а) $f(x) = 2\cdot 1 + (-2)\cdot x + 3 \cdot x^2. \;\;\; f(x) = (2,-2,3)$

б) $f(x) = 3 \cdot x^2 + (-2) \cdot (x - 1) + 0 \cdot 1 \;\;\; f(x) = (3,-2,0)$


__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

**Решение**

а) Нет, не является. Приведем контрпример. Пусть $a = (0,1,3), ~ b = (1,0,5)$. Тогда $ a + b = (1,1,8)$. Такой вектор не принадлежит нашему множеству, а значит оно подпространством не является.

б) Пусть $a = a_1 \cdot u_1 + a_n \cdot u_1 + ... + a_n \cdot u_n, ~ b = b_1 \cdot u_1 + b_2 \cdot u_2 + ... + b_n \cdot u_n $. Тогда

$$a + b = (a_1 + b_1) \cdot u_1 + (a_2 + b_2) \cdot u_2 + ... + (a_n + b_n) \cdot u_n $$

$$\lambda \cdot a = (\lambda \cdot a_1)\cdot u_1 + (\lambda \cdot a_2)\cdot u_2 + ... + (\lambda \cdot a_n)\cdot u_n $$

Так как $a+b, ~\lambda \cdot a$ также принадлежат нашему множеству, то оно является линейным подпространством.

## **Урок 2.**

__1.__ Найти скалярное произведение векторов $x, y \in \mathbb{R}$:
<br>
а) $x=(0,-3, 6),~y=(-4, 7, 9);$
<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

**Решение**

а)  $x \cdot y = 0 \cdot (-4) + (-3) \cdot 7 + 6 \cdot 9 = 40 $
<br>
б)  $x \cdot y = 7 \cdot (-3) + (-4) \cdot 1 + 0 \cdot 11 + 1 \cdot 2 = -23 $

__2.__ Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

**Решение**

$$\|a\| = \sqrt{4^2 + 2^2 + 4^2} = 6,~\|b\| = \sqrt{12^2 + 3^2 + 4^2} = 13, ~ a \cdot b = 4 \cdot 12 + 2 \cdot 3 + 4 \cdot 4 = 70 $$

$$\cos \widehat{(a,b)} = \frac{a \cdot b}{\|a\|\cdot\|b\|} = \frac{35}{39}$$


__3.__ Будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
а) произведение длин векторов;<br>
б) утроенное обычное скалярное произведение векторов?<br>

**Решение**

a) $ a \cdot b = 0 \Leftrightarrow |a| = 0 $ и $|b| = 0 $. Просторанство будет Евклидовым тогда и только тогда, когда оно одномерное.

б) Будет. Ортонормированным базисом в таком пространстве будет набор $\xi_1 = (\frac{1}{\sqrt{3}},0,\dots,0), \xi_2 = (0, \frac{1}{\sqrt{3}},\dots,0), \dots, \xi_n = (0, 0,\dots,\frac{1}{\sqrt{3}}) $

__4.__ Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$ 

**Решение**

а) Не будет. Этот набор не является базисом, так как содержит всего 2 вектора. Например, вектор (0,1,0) через эти два вектора не выражается. Однако, в подпространстве векторов, у которых ворая коррдината равна нулю, это будет базис.
<br>

б) Будет. В самом деле, пусть $a_1 = (\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}},0),~a_2 = (\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}},0),~a_3 = (0,0,1) $. Тогда $ a_1\cdot a_1 = 1, ~ a_2\cdot a_2 = 1, ~ a_3\cdot a_3 = 1, ~ a_1\cdot a_2 = 0, ~ a_1\cdot a_3 = 0, ~ a_2\cdot a_3 = 0$. Значит, базис является ортонормированным.
<br>

в) Не будет. В самом деле, пусть $a_1 = (\frac{1}{2}, -\frac{1}{2},0),~a_2 = (0, \frac{1}{2}, \frac{1}{2}),~a_3 = (0,0,1) $. Тогда $\|a\| = \|b\| = \frac{1}{2} \neq 1, ~ 
a_1 \cdot a_2 = -\frac{1}{4} \neq 0$
<br>

г) Будет. Это очевидно
